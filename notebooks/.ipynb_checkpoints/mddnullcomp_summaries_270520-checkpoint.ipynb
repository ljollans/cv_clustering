```python
def aggregate_mainfolds(input_filedir, modstr, s, minvarexp=1/3):

    with open((input_filedir + 'aggr_betas_best_sf_k.pkl'), 'rb') as f:
        [bestk_mdd,final_k,allbetas_sfk,silmaintrain,silmaintest,train_clus_assig,test_clus_assig,train_clus_prob,test_clus_prob,train_clus_assig25,test_clus_assig25] = pickle.load(f)
 
    filestr=(input_filedir + sets[s] + modstr + str(0))
    with open(filestr, "rb") as f:
        mod = pickle.load(f)
        
    tsne = TSNE(n_components=2, perplexity=30, early_exaggeration=1, learning_rate=2, n_iter=1000).fit_transform(mod.data)
    # preallocate
    newlabel_test = np.zeros(shape=[mod.data.shape[0]])
    cert_test = np.zeros(shape=[mod.data.shape[0]])
    newlabel_train = np.zeros(shape=[mod.data.shape[0]])
    cert_train = np.zeros(shape=[mod.data.shape[0]])
    trainrelab = np.full([mod.data.shape[0],4],np.nan)

    # collect all betas and compare
    tmp=[]
    fig=plt.figure(figsize=[20,7])
    for mf in range(4):
        tmp.append(allbetas_sfk[s][mf])
        print('mainfold ' + str(mf) + ': ' + str(allbetas_sfk[s][mf].shape[1]) + ' clusters')

        maintest = np.where(np.isnan(mod.cv_assignment[:, mf]))[0]
        maintrain = np.where(np.isfinite(mod.cv_assignment[:, mf]))[0]
        
        ax=plt.subplot(2,4,mf+1)
        colorscatter(tsne[maintrain,:],train_clus_assig25[maintrain,s,mf],train_clus_prob[maintrain,s,mf]*4,ax,1)
        plt.title('training set')
        ax=plt.subplot(2,4,mf+5)
        colorscatter(tsne[maintest,:],test_clus_assig25[maintest,s,mf],test_clus_prob[maintest,s,mf]*4,ax,1)
        plt.title('test set')
    plt.show()
    
    allmses, allcorrs, allvarexp = mainfold_clus_comp(tmp,np.max(final_k[s,:]).astype(int)-2)
    print('max variance explained:', np.nanmax(allvarexp))
    sns.distplot(allvarexp.flatten())
    plt.show()
    
    assignedwhere=varexpmatcher(allvarexp, np.max(final_k[s,:]).astype(int)-2, 1/4)
    collected_betas=np.full([mod.data.shape[1],np.nanmax(assignedwhere).astype(int),4],np.nan)
    for mf in range(4):
        maintest = np.where(np.isnan(mod.cv_assignment[:, mf]))[0]
        maintrain = np.where(np.isfinite(mod.cv_assignment[:, mf]))[0]
        # fill the re-labeled assignments from the test sets into one vector
        for c in range(np.max(final_k[s,:]).astype(int)):
            if np.isfinite(assignedwhere[mf,c]):
                newlabel_test[maintest[np.where(test_clus_assig25[maintest,s,mf]==c+1)[0]]]=assignedwhere[mf,c]
                trainrelab[maintrain[np.where(train_clus_assig25[maintrain,s,mf]==c+1)[0]],mf]=assignedwhere[mf,c]
                collected_betas[:,assignedwhere[mf,c].astype(int)-1,mf]=allbetas_sfk[s][mf][:,c]
        cert_test[maintest]=test_clus_prob[maintest,s,mf]
        
    for i in range(mod.data.shape[0]):
        ccount=np.array([len(np.where(trainrelab[i,:]==c)[0]) for c in range(np.max(final_k[s,:]).astype(int)+1)]) # here i am also counting zeros 
        # which indicates no assignment
        maxccount=np.nanmax(ccount)
        maxccountwho=np.where(ccount==maxccount)[0]
        if len(maxccountwho)==1:
            if maxccount>=2:
                newlabel_train[i]=maxccountwho
                cert_train[i]=maxccount
            else:
                cert_train[i]=4-maxccount
        else:
            cert_train[i]=4-maxccount
        
    fig=plt.figure(figsize=[20,5])
    ax=plt.subplot(1,2,1)
    colorscatter(tsne,newlabel_test,cert_test*4,ax,1)
    ax=plt.subplot(1,2,2)
    colorscatter(tsne,newlabel_train,cert_train,ax,1)
    plt.show()

    
    return newlabel_test ,cert_test ,newlabel_train ,cert_train, collected_betas
    
```

```python
def returnanova(var1, var2):
    # anova for slope
    results = ols(var1 + '~ C(' + var2 + ')', data=clin).fit()
    aov_table = sm.stats.anova_lm(results, typ=2)
    p = aov_table['PR(>F)'][0]
    return p
```
